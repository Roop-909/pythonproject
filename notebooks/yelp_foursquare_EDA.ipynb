```python
import pandas as pd

# Load the combined Foursquare and Yelp data
combined_fsq_df = pd.read_csv('combined_foursquare_data.csv')
combined_yelp_df = pd.read_csv('combined_yelp_data.csv')

# Explore the structure and summary statistics of the Foursquare data
print("Foursquare Data EDA:")
print(combined_fsq_df.info())  # Display data types and non-null counts
print(combined_fsq_df.describe())  # Display summary statistics

# Explore the structure and summary statistics of the Yelp data
print("\nYelp Data EDA:")
print(combined_yelp_df.info())  # Display data types and non-null counts
print(combined_yelp_df.describe())  # Display summary statistics

# Find the number of unique stations in each dataset
unique_fsq_stations = combined_fsq_df['Station Name'].nunique()  # Updated column name
unique_yelp_stations = combined_yelp_df['Station Name'].nunique()  # Updated column name

print(f"\nNumber of Unique Stations in Foursquare Data: {unique_fsq_stations}")
print(f"Number of Unique Stations in Yelp Data: {unique_yelp_stations}")

# Find the top 10 stations with the most available bikes in Foursquare data
top_10_fsq_stations = combined_fsq_df.sort_values(by='Num_Bikes', ascending=False).head(10)  # Updated column name
print("\nTop 10 Stations with Most Available Bikes (Foursquare):")
print(top_10_fsq_stations[['Station Name', 'Num_Bikes']])  # Updated column name

# Find the top 10 stations with the most available bikes in Yelp data
top_10_yelp_stations = combined_yelp_df.sort_values(by='Num_Bikes', ascending=False).head(10)  # Updated column name
print("\nTop 10 Stations with Most Available Bikes (Yelp):")
print(top_10_yelp_stations[['Station Name', 'Num_Bikes']])  # Updated column name

```
